## Bar Graph

### Making a plot from dictionary 

```python
prices={
    "Almond_butter":10,
    "Peanut_butter":8,
    "Cashew butter":12
}
fig,ax=plt.subplots()
ax.bar(prices.keys(),prices.values())
ax.set(title="Butter Price", ylabel="price")
```

## histogram

# Two options for subplot 

```python
#1
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(prices.keys(),prices.values())
ax4.hist(np.random.randn(1000))
```

```python
#2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5));

ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(prices.keys(),prices.values())
ax[1,1].hist(np.random.randn(1000))
```

```python
heart_disease_data= pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/refs/heads/master/data/heart-disease.csv");
heart_disease_data
```

```python
heart_disease_data.head()
```

```python
heart_disease_data["age"].plot.hist()
```

```python
heart_disease_data.plot.hist(subplots=True,figsize=(10,30));
```

```python
over_50_data=heart_disease_data[heart_disease_data["age"]>50];
over_50_data
```

```python
##pyplot
over_50_data.plot(kind='scatter',
                 x='age',y='chol',c='target')
```

```python
## OO method
fig,ax=plt.subplots(figsize=(10,6)); 
scatter=ax.scatter(over_50_data["age"],over_50_data["chol"],c=over_50_data["target"],cmap="viridis");
ax.set_xlabel("Age")
ax.set_ylabel("Cholesterol")
ax.set_title("Age vs Cholesterol levels (Colored by Target)")
# ax.set_xlim([45,100])

#add legend
ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(over_50_data["chol"].mean(),linestyle="--")
```

```python
over_50_data.head()
```

```python
# subplot of chol,age,thalach

fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10));

#add data to ax0

scatter=ax0.scatter(x=over_50_data["age"],y=over_50_data["chol"],c=over_50_data["target"]);
ax0.set(title="Heart Disease and Cholestrol Level",xlabel="Age",ylabel="Cholestrol")

ax0.legend(*scatter.legend_elements(),title="Target")

ax0.axhline(y=over_50_data["chol"].mean(),linestyle="--")

#add data to ax1

scatter_1=ax1.scatter(x=over_50_data["age"],y=over_50_data["thalach"],c=over_50_data["target"])
ax1.set(title="Heart Diseases and Thalach" ,xlabel="age",ylabel="thalach" )

ax1.legend(*scatter_1.legend_elements(),title="Target");
ax1.axhline(y=over_50_data["thalach"].mean(),linestyle="--")

fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
